```python
import numpy as np
import pandas as pd
```

```python
import tqdm
import glob, os, pickle
dfs=[]
path= "/dartfs/rc/nosnapshots/V/VaickusL-nb/EDIT_Students/users/Sophie_Chen/scc_tumor_data/prelim_patch_info_v2/"

for file in tqdm.tqdm(sorted(glob.glob(path + '/*.pkl'))):
    basename=os.path.basename(file).replace("_ASAP_tumor_map.pkl","")
    df=pd.read_pickle(file)
    # we want to set y_true = 0,1,2 corressponding to benign, scc, inflamm
    print(df.groupby("y_true").size())
    df.loc[df.inflamm == 1, "y_true"] = 2
    df.loc[df.scc == 1, "y_true"] = 1
    print(sum(df["y_true"]==1), sum(df["y_true"]==2), sum(df["y_true"]==0))
    df['patch_index']=np.arange(df.shape[0])
#     if 'scc' not in df.columns: df['scc']=0
#     df['scc']=(df['scc']==1).astype(int)  
#     if 'inflamm' not in df.columns: df['inflamm']=0
#     df['inflamm']=(df['inflamm']==1).astype(int)  
    df['ID']=basename
    dfs.append(df)
df=pd.concat(dfs)
```

```python
train_set=pd.DataFrame(file_info[file_info.set=='train'])
val_set=pd.DataFrame(file_info[file_info.set=='val'])
test_set=pd.DataFrame(file_info[file_info.set=='test'])
```

```python
train_val_test_slide_ids=dict(train=train_set.IDs,val=val_set.IDs,test=test_set.IDs)
```

```python
np.random.seed(42)
patch_dict=dict(train=df.loc[df['ID'].isin(train_val_test_slide_ids['train'])],
               val=df.loc[df['ID'].isin(train_val_test_slide_ids['val'])],
               test=df.loc[df['ID'].isin(train_val_test_slide_ids['test'])])
```

```python
patch_dict
```

```python
train_val_test_slide_ids={k:patch_dict[k]['ID'].unique() for k in patch_dict}
```

```python
train_val_test_slide_ids
```

```python
df['Set']='train'
for k in train_val_test_slide_ids:
    df.loc[df["ID"].isin(train_val_test_slide_ids[k]),"Set"]=k
```

```python
import tqdm, os, glob, pickle 
for k in tqdm.tqdm(list(patch_dict.keys()),desc="train_val_test"):
    if not os.path.exists(f'cnn_model_input/{k}_data.pkl'):
        X,y=[],[]
        for name, dff in tqdm.tqdm(patch_dict[k].groupby('ID'),total=patch_dict[k]['ID'].nunique(),desc="case"):
            arr=np.load(f"/dartfs/rc/nosnapshots/V/VaickusL-nb/EDIT_Students/users/Sophie_Chen/scc_tumor_data/prelim_patch_info/{name}_ASAP_tumor_map.npy")
            X.append(arr[dff['patch_index'].values])
            y.append(dff['y_true'].values.flatten().astype(int))
            del arr
        X=np.concatenate(X,0)
        y=np.hstack(y)
        with open(f'/dartfs/rc/nosnapshots/V/VaickusL-nb/EDIT_Students/users/Sophie_Chen/scc_tumor_data/prelim_patch_info/{k}_data.pkl','wb') as f:
            pickle.dump(dict(X=X,y=y,patch_info=patch_dict[k]),f, protocol = 4) #pickle.HIGHEST_PROTOCOL)
```

```python
import sklearn.metrics as metrics
# fpr and tpr of all thresohlds
true = predictions['true']
preds = predictions['pred'][:,1]
fpr, tpr, threshold = metrics.roc_curve(true, preds)
roc_auc = metrics.auc(fpr, tpr)

```

```python

import matplotlib.pyplot as plt
%matplotlib inline
plt.title('AUC-ROC')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
